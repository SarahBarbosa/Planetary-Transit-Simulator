# Importando a biblioteca

```python
from SimTransP import SimTransP as STP
```

# Criando uma curva de luz

```python
# Vamos criar uma instância STP com duração de 90 dias e uma cadência de 0.020833 (Kepler mission)
cl = STP(tmin = 0, tmax = 90, cadencia = 0.020833, ai = 90)

#A instância cl corresponde à curva de luz de uma estrela de disco uniforme
#vamos plotar para ver isso
cl.plot()
```

# Adicionando spots e fáculas

```python
# Precisamos chamar o add_spots e definir o período de rotação da estrela
cl.add_spots(Peq=30)

# e plotar novamente
cl.plot(titulo = 'Curva de luz com spots e fáculas')
```

# Adicionando ruído branco

```python
# Definindo um desvio padrão
std = 1000e-6

# Chamando o atributo add_noise
cl.add_noise(std = std)

# e plotando
cl.plot(titulo = 'Curva de luz spots + ruído')
```

# Adicionando trânsito planetário

```python
# Precisamos especificar:
# - raio da estrela Re em raios solares
# - o raio do planeta Rp em razão do raio solar
# - Período orbital do planeta
cl.add_transito(Re = 1, Rp = 0.15, Porb = 10)

# Re-plotando o gráfico após o trânsito planetário
cl.plot(titulo = 'Curva de luz com spots + ruído + trânsito planetário')
```

# Periodograma lomb scargle

```python
# Podemos calcular o periodograma da estrela chamando apenas periodograma()

cl.periodograma()
```
